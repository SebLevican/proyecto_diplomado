## 2 **Transformación y Creación de indicadores**

### 2.1 Aplicación logaritmo

### 4.2 Test Chi-square

```python
import pandas as pd
from scipy.stats import chi2_contingency

str_columns = data.select_dtypes(include=['object'])
alpha = 0.05

for columna in str_columns.columns:
    c_table = pd.crosstab(data['FUGA'], data[columna])
    stat, p_val, dof, expected = chi2_contingency(c_table)

    print(f"Análisis de la columna: {columna}")
    print(f"p-value = {round(p_val,4)} | estadístico: {round(stat,3)}")
    if p_val <= alpha:
        print('Las variables son dependientes (Rechaza Hipótesis Nula)  | Existe relación entre las variables')
    else:
        print('Las variables son independientes (Acepta Hipótesis Nula) | No existe relación entre las variables')
    print("-------------------------------------------")
```

Como se rechaza la hipótesis nula, las columnas categorias son relevantes al momento de predecir si un cliente se fugara.

```python
def cramers_v(contingency_table):
    chi2, _, _, _ = chi2_contingency(contingency_table)
    n = contingency_table.sum().sum()
    k = min(contingency_table.shape) - 1
    return np.sqrt(chi2 / (n * k))

results = {}

for col in ['GENERO', 'NIV_EDUC', 'E_CIVIL', 'CIUDAD', 'SEGURO']:
    contingency_table = pd.crosstab(data[col], data['FUGA'])
    results[col] = cramers_v(contingency_table)

# Ordenar y mostrar los resultados
sorted_results = sorted(results.items(), key=lambda x: x[1], reverse=True)
for col, v in sorted_results:
    print(f"{col}: Cramer’s V = {v:.4f}")
```

A continuación, se calculó el coeficiente de Cramér's V para evaluar la fuerza de asociación entre la variable objetivo (FUGA) y varias variables categóricas. Los resultados muestran los valores de Cramér's V para cada variable:

    CIUDAD: Cramér’s V = 0.5895
    NIV_EDUC: Cramér’s V = 0.5122
    GENERO: Cramér’s V = 0.4744
    E_CIVIL: Cramér’s V = 0.2850
    SEGURO: Cramér’s V = 0.0812

Siguiendo el criterio de seleccionar variables con un Cramér’s V superior a 0.5, se eligen las variables CIUDAD y NIV_EDUC para el entrenamiento del modelo. Estas variables presentan una asociación más fuerte con la variable objetivo y, por lo tanto, podrían aportar mayor información para la predicción de la fuga.

### 4.3 Test Kolmogorov-Smirnov

```python
def ecdf(data):
    # Computa la distribución empirica acumulada (ECDF por sus siglas en ingles)
    x = np.sort(data)
    n = x.size
    y = np.arange(1, n+1) / n
    return x,y

def plot_kolmogorov(sample1, sample2,variable_name='',alpha=0.05):
  import matplotlib.pyplot as plt
  from scipy import stats
  # obtenemos la ECDF para cada muestra y las ordenamos
  x1,y1 = ecdf(sample1)
  x2,y2 = ecdf(sample2)
  # plot
  plt.figure(figsize=(6, 5))
  sns.lineplot(x=x1, y=y1, color='red')
  sns.lineplot(x=x2, y=y2, color='blue')
  # test
  stat, p_val = stats.ks_2samp(sample1, sample2)
  # interpretamos el p-value
  print(f"p-value: {round(p_val,4)} | alpha: {alpha}")
  if p_val < alpha:
      print('Las distribuciones son independientes (Rechaza Hipótesis Nula) | Existe relación entre las varibles')
  else:
      print('Las distribuciones son dependientes (Acepta Hipótesis Nula) | No existe relación entre las varibles')
  plt.ylabel("Cumulative Distribution Function")
  plt.xlabel(variable_name)
  plt.show()
```

```python
benefit_matrix = {
    'TP': 1000,   # Fuga predicha correctamente (Verdadero positivo)
    'FN': -5000,  # Fuga no predicha (Falso negativo)
    'FP': -100,   # No Fuga predicha como Fuga (Falso positivo)
    'TN': 0       # No Fuga predicha correctamente (Verdadero negativo)
}

# Función para calcular el beneficio total basado en la matriz de confusión
def calcular_beneficio(conf_matrix):
    TP = conf_matrix[1, 1]  # Fuga predicha correctamente
    FN = conf_matrix[1, 0]  # Fuga no predicha
    FP = conf_matrix[0, 1]  # No Fuga predicha como Fuga
    TN = conf_matrix[0, 0]  # No Fuga predicha correctamente

    beneficio_total = (
        TP * benefit_matrix['TP'] +
        FN * benefit_matrix['FN'] +
        FP * benefit_matrix['FP'] +
        TN * benefit_matrix['TN']
    )
    return beneficio_total

# Predicciones de cada modelo
y_pred_log_reg = log_reg.predict(X_test)
y_pred_rf = random_forest.predict(X_test)
y_pred_svm = svm.predict(X_test)

# Matrices de confusión para cada modelo
conf_matrix_log_reg = confusion_matrix(y_test, y_pred_log_reg)
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)
conf_matrix_svm = confusion_matrix(y_test, y_pred_svm)

# Calcular el beneficio para cada modelo usando la función definida anteriormente
beneficio_log_reg = calcular_beneficio(conf_matrix_log_reg)
beneficio_rf = calcular_beneficio(conf_matrix_rf)
beneficio_svm = calcular_beneficio(conf_matrix_svm)

print("Beneficio de Regresión Logística:", beneficio_log_reg)
print("Beneficio de Random Forest:", beneficio_rf)
print("Beneficio de SVM:", beneficio_svm)

# Importancia de características para el modelo Random Forest
importances = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': random_forest.feature_importances_
})
importances = importances.sort_values(by='Importance', ascending=False)


sns.barplot(x='Importance', y='Feature', data=importances.head(10))  # Mostrar las 10 más importantes
plt.title("Características Más Importantes (Random Forest)")
plt.show()
```

```python
# Creamos indicadores, nuevas variables en el dataset limpio que nos pueden ayudar a entener tendencias

# 1. Promedio de deudas mensuales
data_validacion['Promedio_Deuda'] = data_validacion[['D_Marzo', 'D_Abril', 'D_Mayo', 'D_Junio', 'D_Julio', 'D_Agosto', 'D_Septiembre']].mean(axis=1)

# 2. Ratio de morosidad (número de veces en mora / número de meses con deuda)
months_with_debt = (data_validacion[['D_Marzo', 'D_Abril', 'D_Mayo', 'D_Junio', 'D_Julio', 'D_Agosto', 'D_Septiembre']] > 0).sum(axis=1)
data_validacion['Ratio_Morosidad'] = data_validacion['M_MOROSO'] / months_with_debt.replace(0, 1)

# 3. Renta por Edad
data_validacion['Renta_Edad'] = data_validacion['RENTA'] / data_validacion['EDAD'].replace(0, 1)

# 4. Endeudamiento Total (Suma de las deudas mensuales)
data_validacion['Endeudamiento_Total'] = data_validacion[['D_Marzo', 'D_Abril', 'D_Mayo', 'D_Junio', 'D_Julio', 'D_Agosto', 'D_Septiembre']].sum(axis=1)

# 5. Deuda respecto a Renta
data_validacion['Deuda_Renta'] = data_validacion['Endeudamiento_Total'] / data_validacion['RENTA'].replace(0, 1)

# 6. Variabilidad de deudas (Desviación estándar de las deudas mensuales)
data_validacion['Variabilidad_Deuda'] = data_validacion[['D_Marzo', 'D_Abril', 'D_Mayo', 'D_Junio', 'D_Julio', 'D_Agosto', 'D_Septiembre']].std(axis=1)

# Mostrar las primeras filas con las nuevas variables
data_validacion[['Promedio_Deuda', 'Ratio_Morosidad', 'Renta_Edad', 'Endeudamiento_Total', 'Deuda_Renta', 'Variabilidad_Deuda']].head()
```
